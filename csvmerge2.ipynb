```python
import pandas as pd 

df1 = pd.read_csv("전남_100까지.csv")       # 합칠 파일 경로 1

df = pd.read_csv("전남100p~.csv")


result = pd.concat([df1,df])
result.drop('Unnamed: 0',axis=1,inplace=True)           # 합치고 나서 이상한 컬럼 생겨서 제거
result.drop_duplicates(inplace=True)                    # 중복 제거
result.reset_index(drop=True,inplace=True)              # 중복 제거 후 인덱스 재설정
result.to_csv('전남.csv')                            # csv 파일로 저장
```
